# Employee Attrition - Visualizing the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
employee_data = pd.read_csv("data/prepared-employee_attrition.csv")
employee_data
```

```python
plt.figure(figsize=(13,9))
sns.heatmap(employee_data.corr(),vmax=0.8,linewidth=0.1,cmap='coolwarm')
plt.show()
```

```python
employee_data['Attrition'].value_counts().plot(kind='pie')
```

```python
plt.figure()
sns.displot(employee_data, x="Age", )
plt.show()
```

```python
plt.figure()

pairgrid = sns.PairGrid(data=employee_data, y_vars=["Age","DistanceFromHome","MonthlyIncome","MonthlyRate","TotalWorkingYears", "YearsAtCompany"], x_vars=["Age","DistanceFromHome","MonthlyIncome","MonthlyRate","TotalWorkingYears", "YearsAtCompany"], hue="Attrition")
pairgrid.map_diag(sns.histplot, bins=20, kde=True)
pairgrid.map_upper(sns.scatterplot)
pairgrid.map_lower(sns.kdeplot, legend=True)
plt.show()
```

```python
plt.figure()
sns.histplot(data=employee_data, x="JobSatisfaction", hue="Attrition")
plt.show()
```

```python
plt.figure()
sns.histplot(data=employee_data, x="MonthlyIncome", hue="Attrition")
plt.show()
```

```python
plt.figure()
sns.kdeplot(data=employee_data, x='MonthlyIncome', y="WorkLifeBalance",hue='Attrition', fill="true", alpha=0.7)
plt.show()
```

```python
plt.figure()
sns.kdeplot(data=employee_data, x='MonthlyIncome', y="JobSatisfaction", hue='Attrition', fill="true", alpha=0.7)
plt.show()
```

```python
plt.figure()
sns.histplot(data=employee_data, x="OverTime", hue="Attrition")
plt.show()
```
